```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
iris = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/IRIS.csv')
```

```python
iris.head(6)
```

```python
iris.columns
```

```python
iris.info()
```

```python
iris.isnull().sum()
```

```python

for i in iris.columns:
   if(i!='species'):
       plt.figure(figsize=(3,2))
       sns.scatterplot(x=iris[i],y=iris['species'])
       plt.show()
   else:
         break
```

```python
x=iris[['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
y=iris['species']
```

```python
plt.plot(iris.iloc[:,:-1],y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=2529)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1=LogisticRegression(multi_class='ovr')
```

```python
model1.fit(x_train,y_train)
```

```python
model2=LogisticRegression(multi_class='multinomial')
```

```python
model2.fit(x_train,y_train)
```

```python
model3=LogisticRegression()
```

```python
model3.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)*100
```

```python
model2.score(x_test,y_test)*100
```

```python
model3.score(x_test,y_test)*100
```

```python
#prediction
y_pred1=model1.predict(x_test)
```

```python

y_pred2=model2.predict(x_test)
```

```python

y_pred3=model3.predict(x_test)
```

```python
x_test.columns
```

```python
y_test
```

```python

plt.figure(figsize=(4,3))
plt.plot(x_test,y_pred1)
plt.title("using One vs Remaining (OvR) technique")
plt.show()
plt.figure(figsize=(4,3))
plt.plot(x_test,y_pred2)
plt.title("using multinomial technique")
plt.show()
plt.figure(figsize=(4,3))
plt.plot(x_test,y_pred3)
plt.title("default technique")
plt.show()
```

```python
print("intercepts with OvR strategy : ",model1.intercept_,"\n")
print("intercepts with multinomial strategy : ",model2.intercept_,"\n")
print("intercepts with default strategy : ",model3.intercept_)
```

```python
print("coefficients with OvR strategy : ",model1.coef_,"\n")
print("coefficients with multinomial strategy : ",model2.coef_,"\n")
print("coefficients with default strategy : ",model3.coef_,"\n")
```

```python
accuracy_score(y_test,y_pred1)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
accuracy_score(y_test,y_pred3)
```
